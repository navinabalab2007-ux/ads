```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv("Advertising Budget and Sales.csv")
```

```python
df.columns
```

```python
df.info()
```

```python
x=df[['TV Ad Budget ($)']]
y=df[['Sales ($)']]
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
plt.scatter(x,y)
plt.xlabel("tv budget")
plt.ylabel("sales")
plt.title("budget vs sales")
plt.show()
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_p=model.predict(x)
```

```python
rmse=np.sqrt(mean_squared_error(y,y_p))
r2=r2_score(y,y_p)
print("RMSE",rmse)
print("R2",r2)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,y_p,color="red")
plt.xlabel("tv budget")
plt.ylabel("sales")
plt.title("budget vs sales")
plt.show()
```
